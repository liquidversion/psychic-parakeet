```python
import matplotlib.pyplot as plt
from quickdraw import QuickDrawDataGroup
import random

# Mode 1: Fixed Logic Mode
# Set fixed doodles that represent your AI stages
concepts = {
    "Research": "book",           # Symbol for knowledge
    "Data": "spreadsheet",        # Represents structured data
    "Testing": "hammer",          # Represents testing/building
    "Feedback": "megaphone"       # Represents loud feedback
}

# Load one drawing for each concept
doodles = [QuickDrawDataGroup(name).get_drawing() for name in concepts.values()]

# Display them with your concept labels
fig, axs = plt.subplots(1, len(doodles), figsize=(15, 4))

for ax, drawing, label in zip(axs, doodles, concepts.keys()):
    ax.imshow(drawing.get_image(), cmap='gray')
    ax.set_title(f"{label}\n({drawing.name.title()})")
    ax.axis('off')

plt.tight_layout()
plt.show()


# Mode 2: Fun, Random Doodles Mode with Logical Representation
# Define possible categories to select from
valid_doodles = [
    "airplane", "alarm clock", "ambulance", "apple", "book", "bridge", "cat", "cloud",
    "cow", "dog", "elephant", "flower", "guitar", "hammer", "house", "light bulb",
    "microscope", "moon", "rainbow", "rocket", "star", "toothbrush", "umbrella"
]

# Randomly pick 4 doodles
random_categories = random.sample(valid_doodles, 4)
random_doodles = [QuickDrawDataGroup(cat).get_drawing() for cat in random_categories]

# Logic for matching doodles to concepts
doodle_concepts = {
    "Data": ["spreadsheet", "bar graph", "pie chart", "database", "calculator"],  # Data-related symbols
    "Testing": ["hammer", "microscope", "toothbrush", "glasses"],  # Tools for testing/inspection
    "Research": ["book", "brain", "notebook", "library"],  # Knowledge/reading
    "Feedback": ["megaphone", "ear", "mouth", "comment", "speaker"],  # Communication and feedback
}

# Display the selected random doodles
fig, axs = plt.subplots(1, len(random_doodles), figsize=(15, 4))

for ax, drawing, category in zip(axs, random_doodles, random_categories):
    # Try to match doodles to AI stages
    matched_category = None
    for key, doodle_list in doodle_concepts.items():
        if drawing.name in doodle_list:
            matched_category = key
            break

    if matched_category is None:
        matched_category = "Unknown"  # If no match found

    ax.imshow(drawing.get_image(), cmap='gray')
    ax.set_title(f"Random Doodle: {drawing.name.title()}\nMapped to {matched_category}")
    ax.axis('off')

plt.tight_layout()
plt.show()
```
